# Throttle Controller

```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
def sigma(x):
    return 1 / (1 + math.exp(-x))

np_sigma = np.frompyfunc(sigma, 1, 1)

_x = np.linspace(-5, 5, 100)
plt.plot(_x, np_sigma(_x))
plt.show()
```

```python
def f_tanh(x):
    return math.tanh(x)

np_f_tanh = np.frompyfunc(f_tanh, 1, 1)

_x = np.linspace(-5,5, 100)
plt.plot(_x, np_f_tanh(_x))
plt.show()
```

```python
def linear(x):
    return x

np_linear = np.frompyfunc(linear, 1, 1)

_x = np.linspace(0, 1, 100)
plt.plot(_x, np_linear(_x))
plt.show()
```

```python
def f_log(x):
    return math.log1p(x)

np_f_log = np.frompyfunc(f_log, 1, 1)

_x = np.linspace(0, 5, 100)
plt.plot(_x, np_f_log(_x))
plt.show()
```
